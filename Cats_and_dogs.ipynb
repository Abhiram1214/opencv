```python
# it's difficult load all the data into the memory because of the size
#so, keras will process it in batches.
# Image generator will help the algo understand an image from various angles

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Activation, Dropout, Conv2D, Flatten, MaxPooling2D, Dense
```

```python
image_gen = ImageDataGenerator(rotation_range=30, 
                               width_shift_range=0.10,
                               height_shift_range=0.1,
                               rescale=1/255,
                               shear_range=0.2, #cropping
                               zoom_range=0.2,
                               horizontal_flip=0.2, #like mirroring; left to right
                               fill_mode = 'nearest')
```

```python
plt.imshow(image_gen.random_transform(cat4))
```

```python
#generate the same for the whole directory
image_gen.flow_from_directory(r'C:\Users\Abhiram Challapalli\Documents\GitHub\opencv\CATS_DOGS\train')
```

```python
input_shape=(150,150,3)
model = Sequential()

model.add(Conv2D(filters=128, kernel_size=(3,3), input_shape=(150,150,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(filters=64, kernel_size=(3,3), input_shape=(150,150,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(filters=64, kernel_size=(3,3), input_shape=(150,150,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())

model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dropout(0.50))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
batch_size = 16

training_image_generator = image_gen.flow_from_directory('CATS_DOGS/train',
                                                                 target_size=(150,150),
                                                                 batch_size=batch_size,
                                                                 class_mode='binary')
```

```python
test_image_generator = image_gen.flow_from_directory('CATS_DOGS/test',
                                                                 target_size=(150,150),
                                                                 batch_size=batch_size,
                                                                 class_mode='binary')
```

```python
training_image_generator.class_indices
```

```python
%%time
results = model.fit(training_image_generator, epochs=100, steps_per_epoch=150, validation_data=test_image_generator,
                   validation_steps=12)
```

```python
plt.plot(results.history['accuracy'])
```
